# TimeSeries Data

Out of all the different datasets, pandas is incredibly useful for TimeSeries data. Classic example time using the Stock Market: https://www.kaggle.com/camnugent/sandp500


```python
import pandas as pd

df = pd.read_csv("./dataset/all_stocks_5yr.csv", parse_dates=["date"], index_col=['Name','date'])
# or remove parse_dates and run
# the second option gives you ability to format the date easier
# df.date = pd.to_datetime(df.date)
df.head()
```

```python
df.describe()
```

```python
df2 = df.sort_index()
df2.head()
```

```python
df2.loc["AAL"]
```

```python
aal = df2.xs("AAL")
```

```python
aal.close.plot();
```

```python
idx = pd.IndexSlice
df_a = df2.loc[idx["A":"AB", :], :]
df_a.close.unstack(level="Name").plot(title="Closing Price");
```

Lets say we only want to look at the the 2017 year. We've seen how to slice on security (ie a string), but what about datetimes? Lets simplify and look at the AAL dataframe so we only have one level of index.
