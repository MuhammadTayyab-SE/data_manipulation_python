# TimeSeries - TimeZones

The bane of any programmers existence. Especially in webdev and databases, you have to be so, so careful. Luckily in this, pandas has your back as well.

To get some time info, lets use some tweets from dear old Elon: https://www.kaggle.com/kulgen/elon-musks-tweets

```python
import pandas as pd

df = pd.read_csv("./dataset/data_elonmusk.csv", encoding='latin1', parse_dates=["Time"])
df = df.drop(df.columns[0], axis=1)
df
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

df.Time.hist(bins=200, grid=False);
```

```python
df = df.rename(columns={"Time": "DateTime"})
df["Time"] = df.DateTime.dt.time
df
```

```python
df.Time.hist(bins=200, grid=False);
```

```python
df.DateTime.dt.hour.value_counts().sort_index().plot.bar();
```

So this probably means the time zone is unset (UTC) and we should put it in Elon's Los Angeles time!So what is this in my time zone? Given Elon is in LA, I assume his time zone is Pacific Standard Time
(GMT -8). As of writing this, I am GMT+10, AEST. So how can I know when I should get on Twitter so Elon can respond to me properly???

First, we tell pandas that the datetime is in fact localized.

```python
df.DateTime = df.DateTime.dt.tz_localize("Europe/Istanbul")
df
```

```python
# df.DateTime.dt.tz_convert(None).dt.tz_localize("America/Los_Angeles")
# Note this doesnt work
```

```python
df["LATime"] = df.DateTime.dt.tz_convert("America/Los_Angeles")
df.LATime.dt.hour.value_counts().sort_index().plot.bar();
```

```python
df["AussieTime"] = df.DateTime.dt.tz_convert("Australia/Brisbane")
df.AussieTime.dt.hour.value_counts().sort_index().plot.bar();
```

```python
import pytz
for t in pytz.all_timezones:
    print(t)
```
