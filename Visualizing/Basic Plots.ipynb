```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
df = pd.read_csv('./dataset/heart.csv')
```

```python
df.head()
```

# Bar Plot

```python
chest_pain = df.groupby('cp').median().reset_index()
chest_pain.head()
```

```python
chest_pain.plot.bar(x = 'cp',y='age')
```

```python
fig,ax = plt.subplots()
ax.bar(chest_pain['cp'], chest_pain['age'], label = 'age')
ax.set_xlabel('cp')
ax.legend()
```

```python
fig,ax = plt.subplots()
ax.bar(chest_pain['cp'], chest_pain['age'], label = 'age',
      edgecolor = 'k', color = ['red','blue','green','black'])
ax.set_xlabel('cp')
ax.legend()
```

```python
ax = sb.barplot('cp', 'age', errcolor = 'b',capsize = 0.1,data = df)
```

# Scatter Plots

```python
df.plot.scatter('age','trestbps')
```

```python
fig,ax = plt.subplots()
ax.scatter(df['age'], df['trestbps'],
      s=60, c=df['age'], alpha=0.5)
ax.set_xlabel('age')
ax.set_ylabel('trestbps')
ax.legend()
```

```python
sb.scatterplot('age','trestbps', hue = 'age', s=30, edgecolor='none',data=df)
```

# Line plots

```python
ages = df.groupby('age').median().reset_index()
ages.head()
```

```python
ages.plot.line('age','chol')
```

```python
ages.plot.line('age',['chol','trestbps'])
```

```python
fig,ax = plt.subplots()
ax.plot(ages['age'], ages['trestbps'],
      ls=':',lw=1.7)
ax.set_xlabel('age')
ax.set_ylabel('trestbps')
```

```python
sb.lineplot('age','trestbps',data=df)
```

```python
df = pd.read_csv('./dataset/heart.csv')
df.head()
```

# Pyplots

```python
fig,axes = plt.subplots(ncols = 2)
df.plot.scatter(x='age', y='chol', ax=axes[0])
df.plot.scatter(x='age', y='trestbps', ax=axes[1])
fig.tight_layout()
```

```python
fig2 = axes[0].get_figure()
```

```python
fig2.savefig('plots.png',bbox_inches='tight',transparant=True)
```

```python
with plt.style.context('default'):
    fig, axes = plt.subplots(ncols =3, sharey=True,
                             gridspec_kw = {'width_ratios': [2,1,1], 'wspace':0})
    y = 'age'
    xs = ['chol','trestbps','thalach']
    for x, ax in zip(xs, axes):
        ax.scatter(df[x], df[y])
        ax.set_xlabel(x)
        axes[0].set_ylabel(y)
```

```python
fig.savefig('output.png',dpi=100,bbox_inches='tight')
```
