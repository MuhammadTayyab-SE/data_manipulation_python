```python
# Code to start you off and manipulate the data. .T is transpose - swap columns and rows
import pandas as pd
df = pd.read_csv("cost-of-living.csv", index_col=0).T.reset_index()
df.head()
```

## Rename column

```python
# your code here
df.rename(columns={"index":"location"}, inplace= True)
```

## Get city and country

```python
df
```

```python
# df.drop(columns='city',inplace=True)
df[['city','country']] = df['location'].str.split(', ',n=1,expand=True)

# df['country'] = df['location'].str.split(', ')[10]
```

## Figure out which country has the most cities

```python
# your code here
most_cities = df.groupby('country')['city'].count().sort_values(ascending=False).index[0]
```

## Create a subset of only that country

```python
# your code here
df2 = df[df['country'] == most_cities]
df2
```

## Sort by housing accommodation

```python
col = "Apartment (1 bedroom) in City Centre"
# your code here
df2.sort_values(col,ascending=False)
df2[['location',col]]
```
