# Pivoting and Margins

Right, lets pick up where we left off.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("happiness.csv")
df.head()
```

```python
df.pivot_table(index="Country", columns="Year", values="Score")
```

```python
df.pivot_table(index="Country", columns="Year", values="Score", margins=True)
```

```python
df.pivot_table(index="Country", columns="Year", values="Score", margins=True,
              aggfunc=np.sum, margins_name="SumOfAll")
```

```python
res = df.pivot_table(index="Country", values=["Score", "GDP per capita"], margins=True,
               aggfunc=[np.mean, lambda x: np.mean(np.diff(x))])
res
```

```python
df.pivot_table(index="Country", columns="Year", values=["Score", "GDP per capita"], margins=True,
               aggfunc=[np.mean, np.median])
```

## Postprocessing data

For fun, lets introduce another function you can use to simplify your data. `qcut`. As in, quantile cut. Cut your data into section based on quantiles. Quantiles being where they sit relative to other data points. So split into top, middle and bottom thirds, for example.

This is very similar to the `cut` method we covered before.

```python
labels = [f"{i + 1} stars" for i in range(5)]
stars = pd.qcut(res[("mean_score", "Score")], 5, labels=labels)
stars.value_counts()
```

```python
res["TrivialisingHappiness"] = stars
res
```
