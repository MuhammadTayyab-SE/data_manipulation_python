# Concats and append

```python
grades1['Semester'] = 1
grades2['Semester'] = 2
grades = pd.concat([grades1,grades2])
grades.head()
```

```python
grades1.copy().append(grades2)
```

```python
grades = grades.reset_index(drop=True)
```

```python
grades
```

# Horizontal merging

```python
student_grades = pd.merge(students, grades,
                          left_on='id', right_on='student_id')
student_grades
```

```python
student2 = students.rename({'id': 'student_id'}, axis=1)
students_full = student2.merge(contacts, on='student_id')
students_full
```

```python
student_grades = pd.merge(student2, grades, on='student_id')
df = pd.merge(student_grades, contacts, on='student_id')
df.head()
```

```python
df2 = df.loc[df.grade == 'F', 'student_id']
method1 = df2.value_counts().rename_axis('student_id')
```

```python
method1.reset_index(name='count')
```

```python
method2 = df.loc[df.grade == 'F',['student_id','grade']].groupby('student_id').count()
method2
```

```python
method2.reset_index()
```

```python
fails = method2[method2['grade'] > 2]
final = students_full[students_full.student_id.isin(fails)]
```
